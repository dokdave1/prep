```python
litr = []
tip = []

for i in df.iloc[:,4]:
    if type(i) == float:
        litr.append(np.nan)
        tip.append(np.nan)
    else:
        litr.append(float(str(i).split(sep=' ')[0]))
        tip.append(str(i).split(sep=' ')[1][1:len(str(i).split(sep=' ')[1])-1])
```

```python
df['Volume'] = np.array(litr)
df['Fuel_type'] = np.array(tip)
```

```python
df = df.drop(columns=['Объем двигателя, л'], axis=1)
df.head()
```

```python
numbers = []
for i in df.iloc[:,5]:
    if type(i) == float:
        numbers.append(0)
    else:
        numbers.append(int(''.join(i.split())[:len(''.join(i.split()))-2]))
```

```python
df['Пробег'] = np.array(numbers)
df.head()
```
